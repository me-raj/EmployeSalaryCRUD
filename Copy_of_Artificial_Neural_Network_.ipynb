 # Q1: Explain cells #Split Data, #Confussion Matrix, and # Show the Accuracy, Precision, Recall, F1, etc. 


1. To split the dataset and create training data and test data, we call the function train_test_split with the following parametres:
 train_feature - input array that contains the feature data
 train_target - output (target) data
 shuffle = True means that the data are shuffled into random order before splitting
 test_size=0.3 means that we split our dataset into 30% test data (and 70% training data)
 random_state=1 means that train_test_split will shuffle the data in the same order prior to the split, every time you use the function with that same integer (1)

2.    accuracy represents the number of correctly classified data instances over the total number of data instances (Accuracy = TP+TN/TP+FP+FN+TN)
3.   precision is the ratio of correctly predicted positive observations to the total predicted positive observations (Precision = TP/TP+FP
4. recall is the ratio of correctly predicted positive observations to the all observations (Recall = TP/TP+FN)
5.  F1 Score is the weighted average of Precision and Recall (F1 = 2*(Recall * Precision) / (Recall + Precision))
6. Confussion Matrix is a predictor where we can measure (or visualize) the performance of our used classification algorithm. 


 the metric used to measure the performance of machine learning classification models based on assessing the perfect agreement and agreement by chance between the two raters (real-world observer and the classification model).





```python
# Import Library
import pandas as pd
```

```python
#Import the files to Google Colab
url = 'https://raw.githubusercontent.com/rc-dbe/bigdatacertification/master/dataset/churn_trasnsformed_new.csv'
df_csv = pd.read_csv(url, sep=',',)

# Show 10 first Row
df_csv.head()
```

```python
# Remove "Unnamed:O" Coloumn
df = df_csv.drop("Unnamed: 0", axis=1)
df.head()
```

```python
# Check the Data Infomation
df.info()
```

```python
#Import MinMax Scaler
from sklearn.preprocessing import MinMaxScaler

# initialize min-max scaler
mm_scaler = MinMaxScaler()
column_names = df.columns.tolist()
column_names.remove('Churn')

# Transform all attributes
df[column_names] = mm_scaler.fit_transform(df[column_names])
df.sort_index(inplace=True)
df.head()
```

```python
# Selecting the Feature of TotalCharges , by remove the unused feature 
feature = ['TotalCharges']
train_feature = df.drop(feature, axis=1)

# Set The Target
train_target = df["TotalCharges"]
```

```python
# Show the Feature
train_feature.head(5)
```

```python
# Split Data
from sklearn.model_selection import train_test_split, cross_val_score
# We import functions train_test_split, cross_val_score from Sklearn's model selection module 
X_train, X_test, y_train, y_test = train_test_split(train_feature ,train_target, shuffle = True, train_size=0.1, test_size=0.3, random_state=1)
```

# Q3: 3. Change the output parameter from Churn to TotalCharges, and run again to find out the mean squared error (MSE) value.

 We have to use a "Regression" alrorithm, NOT 'Classification' algorithm, as this is a regression problem

```python
# Import Library
from sklearn.neural_network import MLPRegressor

# Fitting Model
mlp = MLPRegressor(hidden_layer_sizes=(5), activation = 'relu', solver = 'adam',max_iter= 10000, verbose = True)
mlp = mlp.fit(X_train,y_train)

# Prediction to Test Dataset
y_predmlp = mlp.predict(X_test)
```

```python
print('Number of Layer =', mlp.n_layers_)
print('Number of Iteration =', mlp.n_iter_)
print('Current loss computed with the loss function =', mlp.loss_)
```

# As this is a regression problem, we can Not use confusion matrix. We have to use a regressor measure algorithm. 

```python
# Import the Regressor class
from sklearn.multioutput import MultiOutputRegressor
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error

error_function = mean_squared_error(y_test, y_predmlp)

print("The Mean Squarred Error on test set: {:.4f}".format(error_function))

```
